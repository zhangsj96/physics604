```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

Choose n=11, 21 and 31 and plot three curves in one figure. Please use numpy library and matplotlib.

```python
def f(x, n):
    """
    """
    fx = np.pi/2.
    for i in range(1,n+1):
        fx += 2.*np.sin((2*i-1)*x)/(2*i-1)
    return fx
```

```python
x = np.linspace(-np.pi, np.pi, 6000)
```

```python
plt.figure(figsize=(7.5,4))
plt.rc('font', size=12)
plt.plot(x, f(x, 11//2+1), '-', label='n=11')
plt.plot(x, f(x, 21//2+1), '-.', label='n=21')
plt.plot(x, f(x, 31//2+1), '--', label='n=31')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
# No matter how large the n is, spikes will occur at each corner.
```

```python
from vpython import *
scene = canvas(title='Oscillator',
     width=600, height=300,
     center=vector(0,1,2), forward=vector(0,-0.5,-1), background=color.cyan) # set the camera
scene.caption = """Two balls, CoM's velocity is zero"""
scene.forward = vector(0,-.3,-1)

giant = sphere(pos=vector(-2,0,0), radius=1, color=color.red, 
                interval=1, retain=50)
giant.mass = 2
giant.p = vector(0.1, 0, 0) * giant.mass

dwarf = sphere(pos=vector(4,0,0), radius=0.5, color=color.yellow,
                interval=10, retain=50)
dwarf.mass = 1
dwarf.p = -giant.p

arr=helix(pos=giant.pos,axis=dwarf.pos-giant.pos,radius=0.5,color=color.green)

k  = 0.1
dt = 0.1
while True:
    rate(100)
    r = dwarf.pos - giant.pos
    
    F = k * (r-vector(6,0,0))
    giant.p = giant.p + F*dt
    dwarf.p = dwarf.p - F*dt
    
    giant.pos = giant.pos + (giant.p/giant.mass) * dt
    dwarf.pos = dwarf.pos + (dwarf.p/dwarf.mass) * dt
    
    arr.axis = r
    arr.pos = giant.pos
```
